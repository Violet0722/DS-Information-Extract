```python
import urllib.request
import xml.etree.ElementTree
import pandas as pd
```

```python
df = pd.read_csv("./insight-member-name.csv")
df.tail(10)
```

```python
pids = df["PID"]
pids.head(10)
```

```python
for pid in pids:
    url = "https://dblp.org/pid/"+pid+".xml"
    print(url)
    response = urllib.request.urlopen(url)
    raw_xml = response.read().decode()
    tree = xml.etree.ElementTree.fromstring(raw_xml)
    FileName = tree.attrib.get("name")
    rows = []
    for child in tree.findall("r"):
        for xchild in child.findall('article'):
            author_pid = xchild.findall("author")
            for author_pid in xchild.findall("author"): 
                Author = author_pid.text
                Title = xchild.find("title")
                row = { "Title": Title.text,"PublicationID" : xchild.attrib.get("key"), "AuthorPID" : author_pid.attrib.get("pid"),"AuthorORCID" : author_pid.attrib.get("orcid"),
                      "AuthorName": author_pid.text}
                rows.append(row)
        for xchild in child.findall('inproceedings'):
            author_pid = xchild.findall("author")
            for author_pid in xchild.findall("author"): 
                Author = author_pid.text
                Title = xchild.find("title")
                row = { "Title": Title.text,"PublicationID" : xchild.attrib.get("key"), "AuthorPID" : author_pid.attrib.get("pid"),"AuthorORCID" : author_pid.attrib.get("orcid"),
                      "AuthorName": author_pid.text}
                rows.append(row)
        for xchild in child.findall('proceedings'):
            author_pid = xchild.findall("author")
            for author_pid in xchild.findall("author"): 
                Author = author_pid.text
                Title = xchild.find("title")
                row = { "Title": Title.text,"PublicationID" : xchild.attrib.get("key"), "AuthorPID" : author_pid.attrib.get("pid"),"AuthorORCID" : author_pid.attrib.get("orcid"),
                      "AuthorName": author_pid.text}
                rows.append(row)
        for xchild in child.findall('book'):
            author_pid = xchild.findall("author")
            for author_pid in xchild.findall("author"): 
                Author = author_pid.text
                Title = xchild.find("title")
                row = { "Title": Title.text,"PublicationID" : xchild.attrib.get("key"), "AuthorPID" : author_pid.attrib.get("pid"),"AuthorORCID" : author_pid.attrib.get("orcid"),
                      "AuthorName": author_pid.text}
                rows.append(row)
        for xchild in child.findall('incollection'):
            author_pid = xchild.findall("author")
            for author_pid in xchild.findall("author"): 
                Author = author_pid.text
                Title = xchild.find("title")
                row = { "Title": Title.text,"PublicationID" : xchild.attrib.get("key"), "AuthorPID" : author_pid.attrib.get("pid"),"AuthorORCID" : author_pid.attrib.get("orcid"),
                      "AuthorName": author_pid.text}
                rows.append(row)
        for xchild in child.findall('phdthesis'):
            author_pid = xchild.findall("author")
            for author_pid in xchild.findall("author"): 
                Title = xchild.find("title")
                row = { "Title": Title.text,"PublicationID" : xchild.attrib.get("key"), "AuthorPID" : author_pid.attrib.get("pid"),"AuthorORCID" : author_pid.attrib.get("orcid"),
                      "AuthorName": author_pid.text}
                rows.append(row)
        for xchild in child.findall('mastersthesis'):
            author_pid = xchild.findall("author")
            for author_pid in xchild.findall("author"): 
                Author = author_pid.text
                Title = xchild.find("title")
                row = { "Title": Title.text,"PublicationID" : xchild.attrib.get("key"), "AuthorPID" : author_pid.attrib.get("pid"),"AuthorORCID" : author_pid.attrib.get("orcid"),
                      "AuthorName": author_pid.text}
                rows.append(row)
        for xchild in child.findall('www'):
            author_pid = xchild.findall("author")
            for author_pid in xchild.findall("author"): 
                Author = author_pid.text
                Title = xchild.find("title")
                row = { "Title": Title.text,"PublicationID" : xchild.attrib.get("key"), "AuthorPID" : author_pid.attrib.get("pid"),"AuthorORCID" : author_pid.attrib.get("orcid"),
                      "AuthorName": author_pid.text}
                rows.append(row)
    pd.DataFrame(rows).to_csv("C:\\Users\\lenovo\\Desktop\\Google\\dblp\\python-files\\coauthor_new_with_title\\"+FileName+".csv",index=False,encoding="utf-8_sig")
```
